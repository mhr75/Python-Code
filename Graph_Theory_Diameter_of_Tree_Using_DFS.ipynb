```python
### Important functions for executing our work
def detectDiameter(n):
    global maxCount
    DFS(1, n)
    DFS(x, n)
    return maxCount

def DFS(node, n):
    count = 0
    for i in range(n):
        visited[i] = 0
 
    # visited node increment 1
    dfsTraverse(node, count + 1)

def dfsTraverse(node, count):
    global visited, x, maxCount, adj
    visited[node] = 1
    count += 1
    for i in adj[node]:
        if (visited[i] == 0):
            if (count >= maxCount):
                maxCount = count
                x = i
            dfsTraverse(i, count)


### Code Starts from here !!!

  # print(adj)
  # print(visited)

maxCount = float('-inf')
x = 0
ans = input('Do you want to provide your own input or prefer default?\nIf you want to see default contructed tree type ->y\nOtherwise ->n\n')
if ans == 'y': 
  n = 10
  adj, visited = [[] for i in range(n)], [0 for i in range(n)]
  print('Default Constructed tree is')
  print('       0') 
  print('      / \\')
  print('     1   2')
  print('    /      \\')
  print('   4        3')
  print('  / \      / \\')
  print(' 5   6    7   8')
  print('      \\')
  print('       9')
  #======================
    
  # create undirected edges */
  adj[0].append(1)
  adj[1].append(0)

  adj[0].append(2)
  adj[2].append(0)

  adj[1].append(4)
  adj[4].append(1)

  adj[4].append(5)
  adj[5].append(4)

  adj[4].append(6)
  adj[6].append(4)
  adj[6].append(9)
  adj[9].append(6)



  adj[2].append(3)
  adj[3].append(2)


  adj[3].append(7)
  adj[7].append(3)
  adj[3].append(8)
  adj[8].append(3)
else:
  n = int(input('How many Node you want to give input? -'))
  adj, visited = [[] for i in range(n)], [0 for i in range(n)]
  edge = int(input('How many edges are there? -'))
  for i in range(edge):
    a,b = input('Write edge connections giving a space -').split(' ')
    adj[int(a)].append(int(b))
    adj[int(b)].append(int(a))
# print(adj)
print('==================================')
print ("Diameter of the given tree is ", detectDiameter(n))
```
